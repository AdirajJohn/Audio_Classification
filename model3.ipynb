##### Feature Extraction of mfcc

```python
mfcc=[]
label=[]
#1. Zero Crossing
#2. Spectral centroid
zc=[]
sc=[]
#3. Spectral Rolloff
sro=[]
#4. RMS Energy
rms=[]
```

```python

for i in range(1,256):
    #filename=f"esc_{i}"
    filename=f"fsd_{i}"
    # Load audio file and generate spectrogram
    y, sr = librosa.load('D:/data/classification_data/data/coughs/'+filename+".wav")
    mfccs_features = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)

    #extract the feature
    spectral_centroid = (librosa.feature.spectral_centroid(y=y, sr=sr)).mean()
    zero_crossing_rate = (librosa.feature.zero_crossing_rate(y)).mean()
    spectral_rolloff = (librosa.feature.spectral_rolloff(y=y, sr=sr)).mean()
    rms_energy = (librosa.feature.rms(y=y)).mean()
    flag=1

    #Append the result
    sc.append(spectral_centroid)
    zc.append(zero_crossing_rate)
    rms.append(rms_energy)
    sro.append(spectral_rolloff)
    mfcc.append(mfccs_scaled_features)
    label.append(flag)
print(mfcc)
```

```python
x = pd.DataFrame({'mfcc': mfcc})
print(x)
```

```python
x_main=x
```

```python
# Convert the 'Column' column into separate columns
df_features = pd.DataFrame(x['mfcc'].tolist(), columns=['F1',"F2","F3","F4","F5",'F6',"F7","F8","F9","F10",'F11',"F12","F13","F14","F15",'F16',"F17","F18","F19","F20",'F21',"F22","F23","F24","F25",'F26',"F27","F28","F29","F30",
                                                           'F31',"F32","F33","F34","F35",'F36',"F37","F38","F39","F40" ])
```

```python
df_features["Zero_crossing"]=zc
df_features['spectral_centriod']=sc
df_features['rms_energy']=rms
df_features['spectral_rolloff']=sro
```

```python
from sklearn.preprocessing import StandardScaler
# Initialize StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(df_features)
x_scaled_df = pd.DataFrame(x_scale, columns=df_features.columns)
print(x_scaled_df)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
print(len(label))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_scaled_df, label, test_size=0.2, random_state=42)

# Train Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict on test data
y_pred = clf.predict(X_test)

# Evaluate classifier performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
t_a=clf.score(X_train,y_train)
print("Accuracy:", accuracy)
print("Training Accuracy",t_a)
print("Classification Report:\n", report)
```
